# Tokenizer with Encoder and Decoder 

```python
class SimpleTokenizerV1:
    def __init__(self,vocabulary):
        self.str_to_int =  vocabulary
        self.int_to_str = {i:s for s,i in vocabulary.items()}
        
    def encode(self,text):
        preprocessed = re.split(r'[,.:;?_!"()\']|--|\s',text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self,ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Removing spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])',r'\1',text)
        return text
```

```python
tokenizer = SimpleTokenizerV1(vocabulary)
text = '''Then Krishna and Partha (Arjuna) after having narrated everything unto king Yudhishthira the just, introduced Maya unto him.'''

token_ids = tokenizer.encode(text)
print(token_ids)
```

```python
tokenizer.decode(token_ids)
```

# Modifying tokenizer to handle unknown words (mainly <|unk|> and <|endofthetext|>)

```python
all_words = sorted(list(set(preprocessed_tokens)))
all_words.extend(['<|unk|>','<|endoftext|>'])

vocabulary = {id:token for token,id in enumerate(all_words)}
```

```python
class SimpleTokenizerV2:
    def __init__(self,vocabulary):
        self.str_to_int =  vocabulary
        self.int_to_str = {i:s for s,i in vocabulary.items()}
        
    def encode(self,text):
        preprocessed = re.split(r'[,.:;?_!"()\']|--|\s',text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [item if item in self.str_to_int else '<|unk|>' for item in preprocessed]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self,ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Removing spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])',r'\1',text)
        return text
```

# Creating Input-Target Pairs for LLMs
- Here, A Dataloader is created that fetches the input-target pairs from the text using sliding window approach

```python
with open('Book_1.txt','r',encoding='utf-8') as f:
    raw_text = f.read()
enco_text = tokenizer.encode(raw_text)
print(len(enco_text))
```

-Here defining a context-size , context_size = n refers to that a model is allowed to look at 4 words(or tokens) of a sequence, and predict the next word.

```python
enco_sample = enco_text[3544:]
context_size = 4

x = enco_sample[:context_size]
y = enco_sample[1:context_size+1]

print(f'x ====> {x}')
print(f'y ====>     {y}')
```

```python
for i in range(1, context_size+1):
    encoded = enco_sample[:i]
    desired = enco_sample[i]

    print(f'{encoded} ====> {desired}')
```

```python
for i in range(1, context_size+1):
    encoded = enco_sample[:i]
    desired = enco_sample[i]

    print(f'{tokenizer.decode(encoded)} ====> {tokenizer.decode([desired])}')
```

# Building DataLoader
- to load and process data much more efficiently
Building a data-loader using PyTorch's Datasets and DataLoader classes that fetches Input-Target pairs using sliding window approach.

- Step 1 : Tokenize the entire text
- Step 2 : Use sliding window to break the entire text into chunks of sequences of max_length
-  Step 3 : Return total number of rows in the dataset
- Step 4 : Return a single row from the dataset


```python
! pip3 install torch
```

```python
from torch.utils.data import Dataset , DataLoader

class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

    # Tokenize the entire text
    token_ids = tokenizer.encode(txt, allowed_special={'<|endoftext|>'})

    # Using a sliding window to chunk the book into overlapping sequences of max_length
    for i in range(0, len(token_ids) - max_length, stride):
        input_chunks = txt[i:i+max_length]
        output_chunks = txt[i+1:i+max_length+1]
        self.input_ids.append(torch.tensors(input_chunks))
        self.target_ids.append(torch.tensors(output_chunks))

    def __len__(self):
        return len(self.input_ids)
    
    def __getitem__(self):
        return self.input_ids[idx], self.target_ids[idx]
```

# Code below will use GPTDatasetV1
- (that's a Map-style dataset....check in documentation) to load inputs in batches via a PyTorch DataLoader

- Step 1 : Initialize the tokenizer
- Step 2 : Create Dataset
- Step 3 : drop_last=True drops the last batch if it is shorter than the specified batch_size to prevent loss spikes during training 
- Step 4 : The number of CPU Processes to use for preprocessing.

```python
def create_dataloaderV1(txt, batch_size=4, max_length=256, stride=128, shuffle=True, drop_last=True, num_workers=0):
    
    # Initialize tokenizer 
    tokenizer = tiktoken.get_encoding('gpt2')

    # Create Dataset 
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # Create DataLoader
    dataloader = DataLoader(
        dataset, batch_size=batch_size,
        shuffle=shuffle, drop_last=drop_last,
        num_workers=num_workers
    )
    return dataloader
```
